```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

sns.set(style="white")

# Read a dataset
df=pd.read_csv('validation_output.txt',header=True,index_col=False)

df.columns = ['id']+range(6)
df.index = df['id']
df=df.drop('id',1)
```

```python
#df.sort([0],ascending=[0])
df1=df.ix[(df[1]>=0)&(df[1]<10),[0]]
df2=df.ix[(df[1]>=10)&(df[1]<20),[0]]
df3=df.ix[(df[1]>=20)&(df[1]<30),[0]]
df4=df.ix[(df[1]>=30)&(df[1]<40),[0]]
df5=df.ix[(df[1]>=40)&(df[1]<50),[0]]
df6=df.ix[(df[1]>=50)&(df[1]<70),[0]]
df7=df.ix[(df[1]>=70)&(df[1]<100),[0]]
df8=df.ix[(df[1]>=100)&(df[1]<200),[0]]
df9=df.ix[(df[1]>=200)&(df[1]<400),[0]]
df10=df.ix[(df[1]>=400)&(df[1]<700),[0]]
df11=df.ix[(df[1]>=700)&(df[1]<1200),[0]]

x=np.array([5,15,25,35,45,60,85,150,300,550,950])
xerr=np.array([5,5,5,5,5,10,15,50,100,150,250])
y=np.array([df1.stack().mean(),df2.stack().mean(),df3.stack().mean(),df4.stack().mean(),df5.stack().mean(),df6.stack().mean(),df7.stack().mean(),df8.stack().mean(),df9.stack().mean(),df10.stack().mean(),df11.stack().mean()])
yerr=np.array([df1.stack().std(),df2.stack().std(),df3.stack().std(),df4.stack().std(),df5.stack().std(),df6.stack().std(),df7.stack().std(),df8.stack().std(),df9.stack().std(),df10.stack().std(),df11.stack().std()])



fig=plt.figure()
fig.set_size_inches((10,10))
ax = fig.add_subplot(111)
plt.errorbar(x, y, xerr=xerr, yerr=yerr, fmt='o')
ax.plot([0,1200],[0,1200], "r--",label='good',color='green')
ax.set_ylim([-100,1300])
ax.set_xlim([-100,1350])
plt.xlabel('Pushes in September')
plt.ylabel('Pred. Pushes in September')
for item in ([ax.title, ax.xaxis.label, ax.yaxis.label] +
             ax.get_xticklabels() + ax.get_yticklabels()):
    item.set_fontsize(30)
```
